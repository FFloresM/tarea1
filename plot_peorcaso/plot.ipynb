```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a = pd.read_csv("n", sep='\s*,\s*')
n = list(a["n"])
t = list(a["t"])
print(n)
print(t)
```

```python
b = pd.read_csv("n2", sep='\s*,\s*')
n1 = list(b["n"])
t1 = list(b["t"])
print(n1)
print(t1)
```

```python
c = pd.read_csv("nlogn", sep='\s*,\s*')
n2 = list(c["n"])
t2 = list(c["t"])
print(n2)
print(t2)
```

```python
for i in range(len(t2)):
    print(t1[i]>t2[i])
```

```python
plt.plot(n,t,'bo-', label='O(n)')
plt.plot(n1,t1,'ro-', label="O(n^2)")
plt.plot(n2,t2, 'go-', label="O(nlogn)")
plt.legend()
plt.title("arreglo con valores de peor caso")
plt.ylabel("tiempo (ns)")
plt.xlabel("n")
plt.grid()
plt.savefig("peor.png")
plt.show()
```

```python
plt.plot(n[:4],t[:4],'bo-', label='O(n)')
plt.plot(n1[:4],t1[:4],'ro-', label="O(n^2)")
plt.plot(n2[:4],t2[:4], 'go-', label="O(nlogn)")
plt.legend()
plt.title("arreglo con valores de peor caso")
plt.ylabel("tiempo (ns)")
plt.xlabel("n")
plt.grid()
plt.savefig("zoom.png")
plt.show()
```
